```python
import re #regex

dict_palavras = {}
diretorio = '/content/drive/MyDrive/2023/2 SEMESTRE/Pesquisa Ordenacao/'
conteudo = open(diretorio+'texto.txt')
for linha in conteudo:
  palavras = linha.strip().split(' ')#strip retira o \n do final da linha
  #tamanho = len(palavras)
  numPalavras = 0
  if len(palavras[0])>0:
    tamanho = len(palavras)
    for p in range(tamanho):
      palavra_nova = re.sub('[,.:;()\“\”–]','',palavras[p]).lower()
      if palavra_nova not in dict_palavras:
        dict_palavras[palavra_nova]=palavra_nova
      #print(palavra_nova)
      numPalavras+=1
lista_palavras = list(dict_palavras.values())
print(lista_palavras,len(lista_palavras))
with open(diretorio+'palavras.txt','w') as output:
  for palavra in lista_palavras:
    output.write(palavra+'\n')
output.close()
```
